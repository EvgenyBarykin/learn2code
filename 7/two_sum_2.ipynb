```python
def twoSum(self, numbers: List[int], target: int) -> List[int]:
    second = len(numbers)-1
    first = 0
    while numbers[second] + numbers[first] != target:
        if numbers[second] + numbers[first] < target:
            first+=1
        else:
            second-=1
    return first+1, second+1
        
```
