```python
import time

def timer(func):
    def inner(*args, **kwargs):
        num = 10000
        start = time.process_time()
        for i in range(num):
            func(*args, **kwargs)

        end = time.process_time()

        result = (end - start)
        print(type(result))

        print(f'Выполнение функции заняло {round(result*1000, 3)} мс')

    return inner
```

```python
@timer
def move_zeroes(nums) -> None:
    where_to_move = 0
    curr_pos = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            curr_pos = i
            nums[curr_pos], nums[where_to_move] = nums[where_to_move], nums[curr_pos]
            where_to_move+=1
        curr_pos +=1
```

```python
import numpy as np
arr = np.random.randint(-1, 2, 100)
move_zeroes(nums = arr)
```
